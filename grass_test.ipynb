```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data1 = pd.DataFrame(pd.read_csv("./italian_testdata.csv"))
data1 = data1.rename(columns ={"건물수량(kg/ha)":"건물수량"})
indx = data1[data1['연도'] <= 1999].index
data1 = data1.drop(indx)
```

```python
data1.tail()
```

```python
data1.corr()
```

```python
model = smf.ols(formula = '건물수량 ~ 연평균기온', data = data1)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = '건물수량 ~ 최저기온', data = data1)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = '건물수량 ~ 최고기온', data = data1)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = '건물수량 ~ 고온일수', data = data1)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = '건물수량 ~ 연속고온지속일수', data = data1)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = '건물수량 ~ 강수량', data = data1)
result = model.fit()
result.summary()
```
